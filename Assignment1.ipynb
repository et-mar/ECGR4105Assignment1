```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read data from CSV
df = pd.read_csv("HW1.csv")
type(df)

M = df.shape[0]

df.head()
```

### Cost function implementation:

```python
def cost(theta, X, Y):
    error = X.dot(theta) - Y
    cost = np.sum(np.square(error)) / (2* len(Y))
    return cost
```

### Gradient descent algorithm implementation:

```python
def grad_desc(iterations, theta, learning_rate, X, Y):
    cost_history = np.zeros(iterations)
    for i in range(iterations):
        errors = X.dot(theta) - Y
        partials = errors.dot(X)
        theta -= (learning_rate / len(Y)) * partials
        cost_history[i] = cost(theta, X,Y)
    return theta, cost_history
```

## Problem 1
### For feature X1

```python
Y = df.iloc[:,3].to_numpy()
X = df.iloc[:,0].to_numpy()

# Add a column of ones to X for the constant term
X = X.reshape((len(X), 1))
X_0 = np.ones((len(X),1))
X = np.hstack((X_0, X))

iterations = 200
alpha = 0.1
theta_0 = np.zeros(X.shape[1])

theta, cost_history = grad_desc(iterations, theta_0, alpha, X, Y)

print("Final theta: ", theta)
print("Last 5 iterations cost:  ", cost_history[-5:])

# Plot the cost function history
plt.figure()
plt.plot(range(1,iterations+1), cost_history)
plt.title("Convergence of gradient descent")
plt.ylabel("Cost")
plt.xlabel("Number of iterations of gradient descent")

# Plot the data and the best-fit line
plt.figure()
plt.scatter(X[:,1], Y, marker="+")
plt.plot(X[:,1], X.dot(theta), color="red")
plt.title("Result of gradient descent")
plt.ylabel("Y")
plt.xlabel("X1");
```
